```python
from datetime import datetime

# List all your possible date formats
date_formats = [
    "%Y/%m/%d",     # 2025/12/27
    "%Y.%m.%d",     # 2023.03.09
    "%Y-%m-%d",     # 2025-12-04
    "%b %d, %Y",    # Apr 24, 2023
    "%m/%d/%Y",     # 09/23/2023
]

def parse_flexible_date(val):
    for fmt in date_formats:
        try:
            return datetime.strptime(str(val).strip(), fmt)
        except:
            continue
    return pd.NaT

# Apply to your DataFrame's Date column
df['Date_Clean'] = df['Date'].apply(parse_flexible_date)

# Now you can filter or keep only valid dates, as needed
# Example: Only keep rows with valid dates
df = df[df['Date_Clean'].notna()]
```

```python
import pandas as pd

# Sample exchange rate dictionary (update as needed)
exchange_rates = {
    'USD': 1,
    'SGD': 0.74,
    'GBP': 1.27,
    'JPY': 0.0063,
    'EUR': 1.08,
    'CAD': 0.73,
    'AUD': 0.67,
    'INR': 0.012,
    'KRW': 0.00072,
    'TWD': 0.031  # Taiwan Dollar
}

# Assume df is your dataframe and columns are named 'Spend_Cleaned' and 'Currency_Cleaned'
def to_usd(row):
    rate = exchange_rates.get(row['Currency_Cleaned'], None)
    try:
        spend = float(row['Spend_Cleaned'])
    except:
        spend = None  # Handles errors if Spend_Cleaned is not a number
    return spend * rate if (spend is not None and rate is not None) else None

df['Spend_USD'] = df.apply(to_usd, axis=1)
```
